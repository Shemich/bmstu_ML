```python
import numpy as np
import matplotlib.pyplot as plt
from geneticalgorithm import geneticalgorithm as ga

# Определение функции для оценки эмпирического риска (среднеквадратичной ошибки)
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# Функция, которую будем аппроксимировать
def target_function(x):
    return 1 / (1 + 25 * x**2)

# Генерация обучающей выборки и контрольной выборки
def generate_training_data(l):
    return np.array([4 * (i / (l - 1)) - 2 for i in range(1, l + 1)])

def generate_validation_data(l):
    return np.array([4 * ((i - 0.5) / (l - 1)) - 2 for i in range(1, l)])

l = 50
training_data = generate_training_data(l)
validation_data = generate_validation_data(l)
y_true = target_function(validation_data)

# Определение функционала качества для генетического алгоритма
def fitness_function(coefficients):
    y_pred = np.polyval(coefficients, validation_data)
    return mean_squared_error(y_true, y_pred)  # Минимизация ошибки

num_coefficients = 21
varbound = np.array([(-10, 10)] * num_coefficients)  # Диапазон значений коэффициентов

algorithm_param = {'max_num_iteration': 500, 'population_size': 200, 'mutation_probability': 0.2,
                   'elit_ratio': 0.02, 'crossover_probability': 0.8, 'parents_portion': 0.3,
                   'crossover_type': 'uniform', 'max_iteration_without_improv': 50}

model = ga(function=fitness_function, dimension=num_coefficients, variable_type='real', variable_boundaries=varbound,
           algorithm_parameters=algorithm_param)
model.run()

best_coefficients = model.output_dict['variable']
best_mse = model.output_dict['function']

# Визуализация процесса обучения по эпохам
plt.plot(model.report)
plt.xlabel('Generation')
plt.ylabel('Best MSE')
plt.title('Genetic Algorithm Training Progress')
plt.grid(True)
plt.show()

print("Best Coefficients:", best_coefficients)
print("Best Test MSE:", best_mse)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
def target_function(coefficients, x, y_true):
    y_pred = np.polyval(coefficients, x)
    error = np.mean((y_pred - y_true)**2)
    return error
```

```python
def optimize_genetic_algorithm(degree, x_train, y_train):
    num_coefficients = degree + 1
    bounds = [(-10, 10) for _ in range(num_coefficients)]  # Границы для коэффициентов

    def genetic_callback(coefficients):
        return target_function(coefficients, x_train, y_train)

    result = minimize(genetic_callback, np.zeros(num_coefficients), bounds=bounds)
    return result.x
```

```python
def true_func(x):
    return 1 / (1 + 25*x**2)
```

```python
l = 50  # Размер выборки
x_train = np.array([4*(i-1)/(l-1) - 2 for i in range(1, l+1)])
y_train = true_func(x_train)

k = 50  # Размер контрольной выборки
x_control = np.array([4*(i-0.5)/(k-1) - 2 for i in range(1, k)])
```

```python
best_degree = None
min_error = float('inf')

for degree in range(1, 21):
    coefficients = optimize_genetic_algorithm(degree, x_train, y_train)
    error = target_function(coefficients, x_control, y_train[:k])
    
    if error < min_error:
        min_error = error
        best_degree = degree
```
