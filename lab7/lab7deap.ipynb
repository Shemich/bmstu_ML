```python
import numpy as np
import matplotlib.pyplot as plt
from deap import base, creator, tools, algorithms

# Определение функции для аппроксимации
def target_function(x):
    return 1 / (1 + 25 * x**2)

# Определение эмпирического риска (MSE)
def empirical_risk(coefficients, x, y):
    y_pred = np.polyval(coefficients, x)
    return np.mean((y - y_pred)**2)

# Обучающая и контрольная выборки
l = 50
x_train = np.array([4 * (i - 1) / (l - 1) - 2 for i in range(1, l + 1)])
x_control = np.array([4 * (i - 0.5) / (l - 1) - 2 for i in range(1, l)])

y_train = target_function(x_train)
y_control = target_function(x_control)

# Создание генетического алгоритма с использованием DEAP
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", np.ndarray, fitness=creator.FitnessMin)

toolbox = base.Toolbox()
toolbox.register("attr_float", np.random.uniform, -1, 1)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=20+1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

def eval_individual(individual):
    coefficients = individual
    mse = empirical_risk(coefficients, x_train, y_train)
    return mse,

toolbox.register("evaluate", eval_individual)
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=0.2, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

population_size = 50
max_generations = 100
pop = toolbox.population(n=population_size)

# Эволюция популяции
fitness_history = []
for gen in range(max_generations):
    offspring = algorithms.varAnd(pop, toolbox, cxpb=0.7, mutpb=0.2)
    fits = toolbox.map(toolbox.evaluate, offspring)
    for fit, ind in zip(fits, offspring):
        ind.fitness.values = fit
    pop = toolbox.select(offspring, k=len(pop))
    
    best_individual = tools.selBest(pop, k=1)[0]
    fitness_history.append(best_individual.fitness.values[0])


# Аппроксимация на контрольной выборке и сохранение графика
coefficients = best_individual
y_pred_control_epoch = np.polyval(coefficients, x_control)
plt.plot(x_control, y_pred_control_epoch, label=f'Эпоха {gen+1}', linestyle='dashed')
plt.scatter(x_control, y_control, label='Контрольная выборка', color='black', marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Аппроксимация на контрольной выборке по каждой эпохе')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from deap import base, creator, tools, algorithms

def evaluate(individual):
    coefficients = individual
    return empirical_risk(coefficients, x_train, y_train),

population_size = 500
generations = 20
population = toolbox.population(n=population_size)

# Вручную записываем значения ошибки в каждой эпохе
fitness_history = []

stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("min", np.min)

for epoch in range(generations):
    population, logbook = algorithms.eaSimple(population, toolbox, cxpb=0.7, mutpb=0.2, ngen=generations, stats=stats, verbose=False)
    best_individual = tools.selBest(population, k=1)[0]
    best_mse = evaluate(best_individual)[0]
    fitness_history.append(best_mse)

# Извлечение итоговых коэффициентов и ошибки
best_individual = tools.selBest(population, k=1)[0]
coefficients = best_individual
final_mse = evaluate(coefficients)[0]

# Визуализация процесса обучения
y_pred_train = np.polyval(coefficients, x_train)
y_pred_control = np.polyval(coefficients, x_control)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_train, y_train, label='Обучающая выборка')
plt.scatter(x_control, y_control, label='Контрольная выборка')
plt.plot(x_train, y_pred_train, label='Аппроксимация на обучающей', linestyle='dashed')
plt.plot(x_control, y_pred_control, label='Аппроксимация на контрольной', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Аппроксимация функции')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, generations + 1), fitness_history, marker='o')
plt.xlabel('Эпохи')
plt.ylabel('Среднеквадратичная ошибка')
plt.title('Процесс обучения по эпохам')
plt.tight_layout()

# Сохранение графиков
plt.savefig('aproximation_and_training_process1.png')  # Замените имя файла по вашему усмотрению
plt.show()

print("Итоговые коэффициенты:", coefficients)
print("Финальная среднеквадратичная ошибка:", final_mse)
```
