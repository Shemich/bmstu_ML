```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the data into a pandas DataFrame
data = pd.read_excel("data/ЛР6_Полные_данные.xlsx", engine='openpyxl')  
```

```python
# Удаление респондентов с отсутствующими значениями в признаках состояния и причины
df_cleaned = data.dropna(subset=['Оценка.социальной.поддержки', 'Индекс.кредитного.оптимизма', 
                               'Чувство.технологического.прогресса', 'Оценка.риска.безработицы', 
                               'Индекс.семьи', 'Ощущаемое.счастье'], how='any')
```

```python
# Step 1: Convert "Шкала Кантрила" categories to numerical values
scale_mapping = {
    'Prospering': 9,
    'Thriving': 8,
    'Blooming': 7,
    'Doing well': 6,
    'Just ok': 5,
    'Coping': 4,
    'Strugglng': 3,
    'Suffering': 2,
    'Depressed': 1,
    'Hopeless': 0,
}
df_cleaned['Ощущаемое.счастье'] = df_cleaned['Ощущаемое.счастье'].map(scale_mapping)

df_cleaned = df_cleaned.drop(columns=['Респондент', 'Сообщество'])  # Replace '...' with column names you want to drop

# Step 3: Remove commas from numeric values
df_cleaned = df_cleaned.apply(lambda x: x.str.replace(',', '').astype(float) if x.dtype == 'object' else x)

# Step 5: Separate the target variable (Prospering) from the rest of the df_cleaned
X = df_cleaned.drop(columns=['Ощущаемое.счастье'])
y = df_cleaned['Ощущаемое.счастье']

# Rest of the analysis remains the same
# Step 1: Find the features most strongly correlated with individual happiness
correlations = X.corrwith(y)
most_correlated_features = correlations.abs().sort_values(ascending=False).index
print("Most correlated features with Ощущаемое.счастье:")
print(correlations[most_correlated_features])

# Step 2: Train a linear regression model to understand the impact of cause variables on state variables
model = LinearRegression()
model.fit(X, y)

# Get the coefficients to understand the impact of each cause variable on individual happiness
coefficients = pd.Series(model.coef_, index=X.columns).sort_values(ascending=False)
print("\nImpact of cause variables on Ощущаемое.счастье:")
print(coefficients)

# Your machine learning model can be used for prediction and analysis
# You can use these insights to build the model and make predictions on new data.

# For example, to predict Prosperity for a new individual based on their cause variables:
new_individual_cause_variables = [100, 5, 4, 12, 30]
predicted_happiness = model.predict([new_individual_cause_variables])
print("\nPredicted Happiness (Ощущаемое.счастье) for the new individual:", predicted_happiness)
```

```python
# Оставляем только респондентов с индивидуальной характеристикой "Неизвестно"
unknown_respondents_data = data[data['Ощущаемое.счастье'] == 'Неизвестно']
```

```python
# Step 1: Select the relevant features for prediction (same as the selected features used during training)
selected_feature_names = ['Количество.членов.семьи', 'Коэффициент.Джини.сообщества',
                          'Охват.беспроводной.связи.в.сообществе,.%', 
                          'Оценка.риска.безработицы', 'Чувство.технологического.прогресса']  # Replace with the actual names of the selected features

# Step 2: Get the data with only the selected features for the unknown respondents
X_unknown_respondents = unknown_respondents_data[selected_feature_names]

# Step 3: Use the trained model to predict happiness for the unknown respondents
predicted_happiness_unknown = model.predict(X_unknown_respondents)

# Step 4: Add the predicted happiness as a new column to the DataFrame containing the unknown respondents data
unknown_respondents_data['Predicted Happiness'] = predicted_happiness_unknown

# Step 5: Print the DataFrame with the predicted happiness for the unknown respondents
print(unknown_respondents_data)
```
